```python
# import libraries
import pandas as pd
import pathlib
import glob
```

```python
# path
base = pathlib.Path(r'C:\Users\ahmedk40\OneDrive - London South Bank University\Data Science Graduate\ONS\final_exercise')
datasets = base/'datasets'
```

```python
# files
files = [dataset for dataset in datasets.iterdir()]
files
```

```python
# CSV files
csv_files = [csv_file for csv_file in datasets.glob('*.csv')]
csv_files
```

```python
# energy inflation df
df_energy = pd.read_csv(csv_files[4], index_col=[0], header=6, parse_dates=[0], infer_datetime_format=True)
df_energy
```

```python
# cpi inflation df
df_cpi = pd.read_csv(csv_files[0], index_col=[0], header=6, parse_dates=[0], infer_datetime_format=True)
df_cpi
```

```python
# housing inflation df
df_housing = pd.read_csv(csv_files[1], index_col=[0], header=6, parse_dates=[0], infer_datetime_format=True)
df_housing
```

```python
# concatenate all dfs
df_all = pd.concat([df_energy, df_cpi, df_housing], axis=1)
df_all
```

```python
# drop na values
df_all.dropna(axis=0, inplace=True)
df_all
```

```python
# chek na
df_all.isna().sum()
```

```python
# all columns
df_all.columns.tolist()
```

```python
# cleaning, only uk relevant data
df_uk = df_all.drop(columns=['Canada','France','Germany','Italy','Japan','US'])
df_uk.rename(columns={'UK': 'EPI', '12 month percentage change ': 'HPI'}, inplace=True) # EPI: ENERGY_PRICE_INFLATION 
df_uk                                                                                   # HPI: HOUSING_PRICE_INFLATION
```

```python
# export
df_uk.to_csv(datasets/'processed/energy_cpi_housing_processed.csv')
```
